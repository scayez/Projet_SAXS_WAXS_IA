# Tests de Format de la base de données

Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Génération de données 

```python
elements = np.array(["Au", "Ag", "Cu", "Fe", "Ni","Co"])
formes = np.array(["sphere", "icosahedre", "cube", "dodecahedre", "sphere","elipsoid"])
taille1 = np.random.uniform(0.1, 10.0, len(elements))
taille2 = np.random.uniform(0.1, 10.0, len(elements))
taille3 = np.random.uniform(0.1, 10.0, len(elements))
q = [np.linspace(0, 1000000, 1024) for _ in range(len(elements))]
intensites = [np.random.uniform(0, 1000000, 1024) for _ in range(len(elements))]
```

Mise des données en forme dans un pandas dataframe

```python
df = pd.DataFrame({
    'element': elements,
    'forme': formes,
    'taille1': taille1,
    'taille2': taille2,
    'taille3': taille3,
    'q': q,
    'intensites': intensites,
})

df.head()
```

Afficher les courbes

```python

# Créer les subplots
fig, ax = plt.subplots(3, 2, figsize=(15, 10))
ax = ax.flatten()

for i in range(len(elements)):
    ax[i].plot(df['q'][i], df['intensites'][i])
    titre = f"{df['element'][i]} - {df['forme'][i]} - Taille1: {df['taille1'][i]:.2f}, Taille2: {df['taille2'][i]:.2f}, Taille3: {df['taille3'][i]:.2f}"
    ax[i].set_title(titre)

# Supprimer les axes inutilisés si len(elements) < nombre de subplots
for j in range(len(elements), len(ax)):
    fig.delaxes(ax[j])

plt.tight_layout()
plt.show()
```
