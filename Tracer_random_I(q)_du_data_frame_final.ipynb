## II) Définition de la fonctions nécessaire pour le tracer random d'un  I(q)

```python

def tracer_ligne_aleatoire(df, colonne_x="q", colonne_y="Intensités"):
    # Choisir une ligne au hasard
    nbr_alea=int(random.uniform(0, len(df)))
    # Extraire les listes des colonnes spécifiées
    x = df[colonne_x].iloc[nbr_alea]
    y = df[colonne_y].iloc[nbr_alea]
    
   
    print(type(x))
    print(len(y))
    print(len(x))
    
    # Tracer les données
    plt.figure(figsize=(8, 6))
    plt.loglog(x, y)
    plt.xlabel(colonne_x)
    plt.ylabel(colonne_y)
    plt.title(f'Tracé de {colonne_y} en fonction de {colonne_x} pour élément  {df["Element"][nbr_alea]} (de distance= {df["Distance"][nbr_alea]} ) de forme {df["Forme"][nbr_alea]} et de la classe {df["Type solide"][nbr_alea]} avec une de taille  {df["nbr_couche"][nbr_alea]} ')
    plt.grid(True)
    plt.show()
```
